#1. Chi-square goodness-of-fit

A statistical hypothesis test that is used to determine if a sample (categorical variable) is representative of a theoretical distribution. A theoretical distribution is our idea about how should data act.

**1. Stating the Hypotheses:**

$H_0:$

$H_a:$


**Example:**

Each day, the teacher greets students "good morning", "good afternoon", and "good evening" on their way to lectures. He should greet the same rate of students in the morning, afternoon and evening given lectures are evenly distributed throughout the day. 

However, he suspects that he is greeting more students in the afternoon, than the morning and evening.

$H_0: p_{morning} = p_{afternoon} = p_{evening}$ 

$H_a$: At least one $p_i$ is not equal 

($P_i$ = proportion of each greeting)

**3. Checking conditions for inference**

*   Random sample
*   Large count
* Independent choices
* The expected value of  sample observations in each category of the variable is at least 5.




**4. Setting the significance level**

As researchers, we often choose significance $\alpha$ levels equal to 0.01, 0.05, or 0.10.

**5. Checking the expected value assumtion for each category**

$Expected_i = n*p_i$

$E_i$= expected frequency count for level of the categorical variable

$n$ = total sample size

$p_i$ = hypothesized proportion of observations in level i.

**Example:**

$Expected_{morning}: 75 * (1/3) = 25$

$Expected_{afternoon}: 75 * (1/3) = 25$

$Expected_{evening}: 75 * (1/3) = 25$

**6. Calculating a chi-statistics**

Finding the difference between the expected choices and the actual choices.

$$\chi^2 =\frac{(O_i-E_i)^2}{E_i}$$

**Example:**

"Are more students greeting the teacher in the afternoon?"

He sampled 3 random days out of the semester equal to, $N = 75$ greetings, and recorded his rate of the correct greetings (i.e., good morning, good afternoon, good evening). He did not record greetings like "hey dude", "what's up, doc", "later...". 

**Observed number (O):** He actually gets "good morning" = 20, "good afternoon" = 35, "good evening" = 20

$$\chi^2 = \frac{(20-25)^2}{25}+ \frac{(35-25)^2}{25}+ \frac{(20-25)^2}{25} $$

$$\chi^2 = 1 + 4 + 1 = 6.00$$

#2. Chi-square test of homogeneity

A statistical hypothesis test used to determine if the counts of a single categorical variable is distributed identically across different populations. 

**1. State the hypotheses**

$H_0:$

$H_a:$


**Example:**
Each day, a teacher leatures students in accouting from different propulations, HA students and HD students. She assumes that their preference in teaching style is the same. However, she decides to ask the students to identify their preference (e.g., lecture, workshop, online) to determine if there are differences accross populations.  

$H_0: HD_{lecture} = HA_{lecture}$

$H_0: HD_{workshop} = HA_{workshop}$

$H_0: HD_{online} = HA_{online}$

$H_a: At \: least \: one \:of \:the\: H_0 \:statements \:are \:false$


**2. State the significance level**

Common significance levels are equal to 0.01, 0.05 or 0.10. The individual researcher decides what makes sense for their analysis. 

**3. Conditions of inference**

These assumptions needs to be met for the test to be appropriate.

*   For each distribution the sampling is random 
*   The response variable is categorical
*   If data is displayed in (population*category levels) than a frequency count for each expected value is at least 5. 

**4. Checking assumptions of expected value for each combination of population and category.**

**Example:**

Expected frequency count is calculated using the following formula:

$Expected_{r,c} = \frac {n_r * n_c}{n}$

Here, we can check the smallest frequency count to check it meets the requirement of five.

$Expected_{1,1} = \frac {250 * 105}{375} = 70$

$Expected_{1,2} = \frac {250 * 120}{375} = 80$

$Expected_{1,3} = \frac {250 * 150}{375} = 100$

$Expected_{2,1} = \frac {125 * 105}{375} = 35$

$Expected_{2,2} = \frac {125 * 120}{375} = 40$

$Expected_{2,3} = \frac {125 * 150}{375} = 50$


Given that the smallest count > 5, we conclude that the expected count assumption is met.

**5. Calculating the chi-square statistics of homogeneity**

Finding the difference between the expected choices and the actual choices for each distribution.

$$\chi^2 =\sum \frac{(O_{r,c}-E_{r,c})^2}{E_{r,c}}$$

**Example:**

$$\chi^2 =\sum \frac{(O_{r,c}-E_{r,c})^2}{E_{r,c}}$$

$$\chi^2 =\frac{(75-70)^2}{70} + \frac{(75-80)^2}{80} + \frac{(100-100)^2}{100} + \frac{(30-35)^2}{35} + \frac{(45-40)^2}{40} + \frac{(50-50)^2}{50}$$

$$\chi^2 = 0.357 + 0.312 + 0 + 0.714 + 0.625 + 0 = 2.008$$


**6. Check the probability for getting this chi-square statistic using a $chi$-distribution critical values table.**

Degrees of freedom $(df) = (r-1) * (c-1)$

In the example, df = (3-1) * (2-1) = 2

The chi-square value with alpha = 0.05 and 2 degrees of freedom is 5.991, since $\chi_2$ = 2.008 is less than 5.991, we cannot reject $H_0$ and the idea that the teaching preference across HA and HD populations is homogeneous. We conclude that there is no difference between HA and HD populations when accounting for their preferences.

**7. Alternatively, calculate the chi-quare test statistic using python**

```python
from scipy.stats import chi2_contingency

data = [[75,75,100],[30,45,50]]

chi2, p, dof, expected = chi2_contingency(data)

print(chi2,p,dof,expected, sep='\n')
```

# 3. Chi-square test of independence

A statistical hypothesis test used to determine whether two categorical or nominal variables are related or not.

**1. Stating the Hypotheses:**

$H_0$: 

$H_a$:

**Example:**

Each day, the teacher greets students on their way to lectures. However, some students don't greet her. She wonders if the rate of student's greetings are dependent of their majors.  

$H_0$: students' decision to greet is independent of their major

$H_a$:students' decision to greet is not independent of their major


**2. Set significance level**

Decide the alpha value. Specifically, the risk you are willing to take on drawing the wrong conclusion.

For example, $\alpha = 0.05$ means the risk of stating that two variables are independent when they are not.



**3. Conditions of inference**

These assumptions needs to be met for the test to be appropriate.

*   A simple random sample 
*   Two variables: both are categorical
*   More than five expected values for each combination of the two variables. To confirm this, we need to know the total counts for each type of combination and the total counts. 



image.png

**4. Checking the assumption of expected values in each category**
After sampling a total of 181 greetings from students of five different majors. We check the assumption of expected frequency count per category.

Expected frequency count is calculated using the following formula:

$Expected_{r,c} = \frac {n_r * n_c}{n}$

**Example:**

Calculating the expected count for each major - greeting combination needs to exceed 5 to meet the assumptions.

$Expected_{1,1}: \frac{25*100}{181} = 13.81$

$Expected_{1,2}: \frac{25*81}{181} = 11.19$

$Expected_{2,1}: \frac{45*100}{181} = 24.86$

$Expected_{2,2}: \frac{45*81}{181} = 20.14$

$Expected_{3,1}: \frac{46*100}{181} = 25.41 $

$Expected_{3,2}: \frac{46*81}{181} = 20.59$

$Expected_{4,1}: \frac{25*100}{181} = 13.81$

$Expected_{4,2}: \frac{25*81}{181} = 11.19$

$Expected_{5,1}: \frac{40*100}{181} = 22.10$

$Expected_{5,2}: \frac{40*81}{181} = 17.90$

Given that the smallest count is above 5, we can conclude that the expected count assumption is met.



**5. Calculating the chi-square test statistics of independence**

Finding the difference between the expected choices and the actual choices to test for independence between counts.

$$ \chi^2 =\sum \frac{(O_{ij}-E_{ij})^2}{E_{ij}}$$

**Example:**

"Is the rate of student's greetings independent of their majors?"  


$$\chi^2 = 4.88$$

**6. Check the probability for getting this chi-square statistic and the degrees of freedom to look up $chi$-distribution critical value in the table.**

Degrees of freedom $(df) = (r-1) * (c-1)$

In the example, df = (5-1) * (2-1) = 4

The chi-square value with alpha = 0.05 and 4 degrees of freedom is 9.488, since 4.87 is less than 9.488, we can **not reject** $H_0$ and the idea that greetings and majors are independent. We can not conclude that there is relationship between major and decision to greet.
